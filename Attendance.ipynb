```python
imgs=[]
classnames=[]
for i in mylist:
    impath=os.path.join(path,i)
    img1=cv2.imread(impath)
    imgs.append(img1)
    classnames.append(i.split('.')[0])
```

```python
print(classnames)
```

```python
def faceencodings(Images):
    encodelist=[]
    for i in imgs:
        img1=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img1)
        # print(face_in_frame)
        face_encode=fr.face_encodings(img1,face_in_frame)[0]
        encodelist.append(face_encode)
    return encodelist
```

```python
encodelist_knownface=faceencodings(imgs)
```

```python
import datetime
```

```python
def markattendace(name):
    f=open('attendace.csv','a')

    now=datetime.datetime.now()
    time=now.strftime('%H:%M:%S')

    line=name+','+time
    f.write(line+'\n')

    f.close()
```

```python
def markattendace(name):
    f=open('Attendance.csv','r+')
    data=f.readlines()

    names=[]
    for line in data:
        entry=line.split(',')
        names.append(entry[0])

    if name not in names:
        now=datetime.datetime.now()
        time=now.strftime('%H:%M:%S')

        f.write(name+','+time+'\n')

    f.close()
```

```python
import pyttsx3
txt_sp=pyttsx3.init()
voices=txt_sp.getProperty('voices')
txt_sp.setProperty('voice',voices[1].id)
txt_sp.setProperty('volume',0.9)
txt_sp.setProperty('rate',150)
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_loc1=fr.face_locations(img1)
    # print(face_loc1)  
    faceencode1=fr.face_encodings(img1,face_loc1)
    for enc_loc,faceloc in zip(faceencode1,face_loc1):
        matches=fr.compare_faces(encodelist_knownface,enc_loc)
        # print(matches)
        face_dis=fr.face_distance(encodelist_knownface,enc_loc)
        # print(face_dist)
        matchindex=np.argmin(face_dis)
        # print(matchindex)
        if matches[matchindex]==True:
            name=classnames[matchindex]
            markattendace(name)
            text=name
            txt_sp.say(text)
            txt_sp.runAndWait()
            # print(name)
        else:
            print('Unknown')
            text='Unknown'
            txt_sp.say(text)
            txt_sp.runAndWait()
            break

    cv2.imshow('Frame',img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
